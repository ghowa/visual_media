```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
vn = pd.read_csv('vn.csv')
vn[[' AU06_r', ' AU12_r', ' AU15_r']].plot(figsize=(20,10))

vn.loc[vn[' face_id']==3]
```

```python
col = vn.loc[: , [" AU06_r", " AU12_r"]]
vn['happyness'] = col.mean(axis=1)

col = vn.loc[: , [' AU01_r', ' AU04_r', ' AU15_r']]
vn['sadness'] = col.mean(axis=1)

col = vn.loc[: , [' AU01_r', ' AU02_r', ' AU05_r', ' AU26_r']]
vn['surprise'] = col.mean(axis=1)

col = vn.loc[: , [' AU01_r', ' AU02_r', ' AU04_r', ' AU07_r', ' AU20_r', ' AU36_r']]
vn['fear'] = col.mean(axis=1)

col = vn.loc[: , [' AU04_r', ' AU05_r', ' AU07_r', ' AU23_r']]
vn['anger'] = col.mean(axis=1)

col = vn.loc[: , [' AU09_r', ' AU15_r', ' AU16_r']]
vn['disgust'] = col.mean(axis=1)

col = vn.loc[: , [' AU12_r', ' AU14_r']]
vn['contempt'] = col.mean(axis=1)

# boolean
col = vn.loc[: , [" AU06_c", " AU12_c"]]
vn['happyness_bool'] = col.mean(axis=1).round()

col = vn.loc[: , [' AU01_c', ' AU04_c', ' AU15_c']]
vn['sadness_bool'] = col.mean(axis=1).round()

col = vn.loc[: , [' AU01_c', ' AU02_c', ' AU05_c', ' AU26_c']]
vn['surprise_bool'] = col.mean(axis=1).round()

col = vn.loc[: , [' AU01_c', ' AU02_c', ' AU04_c', ' AU07_c', ' AU20_c', ' AU36_c']]
vn['fear_bool'] = col.mean(axis=1).round()

col = vn.loc[: , [' AU04_c', ' AU05_c', ' AU07_c', ' AU23_c']]
vn['anger_bool'] = col.mean(axis=1).round()

col = vn.loc[: , [' AU09_c', ' AU15_c', ' AU16_c']]
vn['disgust_bool'] = col.mean(axis=1).round()

col = vn.loc[: , [' AU12_c', ' AU14_c']]
vn['contempt_bool'] = col.mean(axis=1).round()

# rm = vn.rolling(1).mean()

vn[['happyness_bool', 'sadness_bool', 'surprise_bool', 'fear_bool', 'anger_bool', 'disgust_bool', 'contempt_bool']].plot(figsize=(20,10), style='o')
#vn[['happyness', 'sadness']].plot(figsize=(20,10), 

# vn[['happyness_bool']].plot(style='o')
```
